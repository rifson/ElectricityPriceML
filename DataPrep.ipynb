```python
df_forecast = pd.read_csv("df_forecast.csv", sep=";")
```

```python
df_forecast.head()
```

```python
df_forecast = df_forecast.pivot_table(
    index="HourUTC",
    columns=["PriceArea", "ForecastType"],
    values=["ForecastDayAhead", "ForecastIntraday","Forecast5Hour","Forecast1Hour","ForecastCurrent","TimestampUTC","TimestampDK"],
    aggfunc="first")
```

```python
df_forecast
```

```python
df_forecast = df_forecast.reset_index()
```

```python
df_forecast.columns = ['HourUTC'] + [f'{col[0]}_{col[1]}' for col in df_forecast.columns[1:]]
```

```python
df_forecast
```

```python
df_spotprice = pd.read_csv("df_spotprice.csv", sep=";")
```

```python
df_spotprice.head()
```

```python
df_spotprice = df_spotprice.pivot_table(
    index="HourUTC",
    columns="PriceArea",
    values=["SpotPriceDKK", "SpotPriceEUR"],
    aggfunc="first")
```

```python
df_spotprice = df_spotprice.reset_index()
```

```python
df_spotprice.columns = ['HourUTC'] + [f'{col[0]}_{col[1]}' for col in df_spotprice.columns[1:]]
```

```python
df_spotprice.columns = [col.replace('SpotPrice', '') for col in df_spotprice.columns]
```
